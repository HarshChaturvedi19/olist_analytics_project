# Merging all tables into one master table

```python
# ═══════════════════════════════════════════════
# MASTER TABLE CREATION
# We join all tables into one big table for analysis
# This is like doing multiple SQL JOINs
#
# JOIN ORDER (most logical path):
# orders → customers (who placed the order)
# orders → items (what was ordered)
# items  → products (product details)
# items  → sellers (who sold it)
# orders → payments (how it was paid)
# orders → reviews (what was the rating)
# ═══════════════════════════════════════════════
```

```python
print("=" * 60)
print("BUILDING MASTER TABLE")
print("=" * 60)
```

```python
# ── JOIN 1: orders + customers ───────────────────────────────
print("\n--- Step 1: Merge orders + customers ---")

master = orders.merge(
    customers,
    on='customer_id',   # Both tables have 'customer_id' column → match on this
    how='left'          # Keep all orders, attach customer info where available
)
print(f"Shape after orders + customers: {master.shape}")
```

```python
# ── JOIN 2: master + items ───────────────────────────────────
print("\n--- Step 2: Merge with items ---")

master = master.merge(
    items,
    on='order_id',      # Match using order_id
    how='left'
)
print(f"Shape after + items: {master.shape}")
# Notice: shape rows INCREASE here because one order can have multiple items
# Each item becomes its own row → this is expected and correct
```

```python
# ── JOIN 3: master + products ────────────────────────────────
print("\n--- Step 3: Merge with products ---")

master = master.merge(
    products[['product_id', 'product_category_name', 
              'product_category_name_english', 'product_weight_g']],
    # We only select needed columns from products to keep table clean
    on='product_id',
    how='left'
)
print(f"Shape after + products: {master.shape}")
```

```python
# ── JOIN 4: master + sellers ─────────────────────────────────
print("\n--- Step 4: Merge with sellers ---")

master = master.merge(
    sellers,
    on='seller_id',
    how='left'
)
print(f"Shape after + sellers: {master.shape}")

```

```python
# ── JOIN 5: master + payments ────────────────────────────────
print("\n--- Step 5: Merge with payments ---")

# Problem: one order can have multiple payment rows (split payments)
# Solution: aggregate payments first → get one row per order
payments_agg = payments.groupby('order_id').agg(
    total_payment=('payment_value', 'sum'),    # Total amount paid
    payment_type=('payment_type', 'first'),      # Primary payment method
    installments=('payment_installments', 'max')   # Max installments used
).reset_index()
# .groupby('order_id') → group all payment rows for same order
# .agg() → apply different operations to different columns
# .reset_index() → converts 'order_id' from index back to regular column
# Now payments_agg has exactly ONE row per order_id

master = master.merge(
    payments_agg,
    on='order_id',
    how='left'
)
print(f"Shape after + payments: {master.shape}")
```

```python
# ── JOIN 6: master + reviews ─────────────────────────────────
print("\n--- Step 6: Merge with reviews ---")

# Same potential problem: multiple reviews per order
# Take only the latest review per order
reviews_agg = reviews.groupby('order_id').agg(
    review_score=('review_score', 'mean'),         # Average score if multiple reviews
    review_count=('review_score', 'count')          # How many reviews
).reset_index()

master = master.merge(
    reviews_agg,
    on='order_id',
    how='left'
)
print(f"Shape after + reviews: {master.shape}")
```

```python
# ── FINAL MASTER TABLE OVERVIEW ──────────────────────────────
print("\n--- MASTER TABLE OVERVIEW ---")
print(f"Total rows:    {master.shape[0]:,}")
print(f"Total columns: {master.shape[1]}")
print("\n--- Column Names ---")
print(master.columns.tolist())
print("\n--- Null Values in Master ---")
print(master.isnull().sum()[master.isnull().sum() > 0])
# Only show columns that actually have nulls
```

# master table data cleaning 

```python
# ═══════════════════════════════════════════════
# DATA CLEANING
# Fix all issues found during EDA
# ═══════════════════════════════════════════════
```

```python
print("=" * 60)
print("DATA CLEANING")
print("=" * 60)
```

```python
# ── STEP 1: KEEP ONLY DELIVERED ORDERS ───────────────────────
print("\n--- Step 1: Filter to delivered orders only ---")
print(f"Before filter: {master.shape[0]:,} rows")
```

```python
master_clean = master[master['order_status'] == 'delivered'].copy()
print(f"After filter:  {master_clean.shape[0]:,} rows")
```

```python
# We only want completed orders for analysis
# Cancelled, processing orders are noise for business metrics
# .copy() creates an independent copy so original 'master' stays unchanged
# WHY .copy(): Without it, changes to master_clean might affect master }
```

```python
# ── STEP 2: REMOVE ROWS WITH NO DELIVERY DATE ─────────────────
print("\n--- Step 2: Remove rows missing delivery date ---")
master_clean = master_clean.dropna(subset=['order_delivered_customer_date'])
```

```python
print(f"After removing missing delivery dates: {master_clean.shape[0]:,} rows")
```

```python
# dropna() removes rows that have null values
# subset=['col'] → only remove rows where THIS specific column is null
# Other nulls in other columns are allowed at this step
```

```python
# ── STEP 3: REMOVE ROWS WITH NO PRICE ────────────────────────
print("\n--- Step 3: Remove rows with no price ---")
master_clean = master_clean.dropna(subset=['price'])
print(f"After removing missing prices: {master_clean.shape[0]:,} rows")
```

```python
# ── STEP 4: REMOVE NEGATIVE OR ZERO PRICES ───────────────────
print("\n--- Step 4: Remove invalid prices (<=0) ---")
master_clean = master_clean[master_clean['price'] > 0]
# Filter: keep rows where price is strictly positive
print(f"After removing invalid prices: {master_clean.shape[0]:,} rows")
```

```python
# ── STEP 5: REMOVE NEGATIVE DELIVERY DAYS ────────────────────
print("\n--- Step 5: Remove negative delivery days (data errors) ---")
master_clean = master_clean[master_clean['delivery_days'] >= 0]
# Negative delivery days = delivered before order was placed = impossible = error
print(f"After removing negative delivery days: {master_clean.shape[0]:,} rows")
```

```python
# ── STEP 6: FILL MISSING CATEGORY NAMES ──────────────────────
print("\n--- Step 6: Fill missing category names ---")
master_clean['product_category_name_english'] = (
    master_clean['product_category_name_english'].fillna('Unknown')
)
# fillna('Unknown') replaces all NaN values with the text 'Unknown'
# Better than leaving blanks when doing group analysis
```

```python
# ── STEP 7: FILL MISSING REVIEW SCORES ───────────────────────
print("\n--- Step 7: Fill missing review scores with median ---")
median_score = master_clean['review_score'].median()
# .median() finds the middle value — less affected by outliers than mean
master_clean['review_score'] = master_clean['review_score'].fillna(median_score)
print(f"Filled missing reviews with median score: {median_score}")
```

```python
# ── STEP 8: EXTRACT TIME FEATURES ────────────────────────────
print("\n--- Step 8: Add useful time columns ---")

master_clean['order_year']    = master_clean['order_purchase_timestamp'].dt.year
master_clean['order_month']   = master_clean['order_purchase_timestamp'].dt.month
master_clean['order_quarter'] = master_clean['order_purchase_timestamp'].dt.quarter
# dt.quarter → Q1 (Jan-Mar), Q2 (Apr-Jun), Q3 (Jul-Sep), Q4 (Oct-Dec)
master_clean['order_dayofweek'] = master_clean['order_purchase_timestamp'].dt.day_name()
# day_name() → returns 'Monday', 'Tuesday', etc.
master_clean['order_hour'] = master_clean['order_purchase_timestamp'].dt.hour
# Hour (0-23) → useful for finding peak shopping hours
```

```python
# ── STEP 9: REMOVE DUPLICATE ROWS ────────────────────────────
print("\n--- Step 9: Remove duplicate rows ---")
print(f"Before dedup: {master_clean.shape[0]:,} rows")
master_clean = master_clean.drop_duplicates()
print(f"After dedup:  {master_clean.shape[0]:,} rows")
```

```python
 # ── STEP 10: RESET INDEX ─────────────────────────────────────
master_clean = master_clean.reset_index(drop=True)
# After dropping rows, the index has gaps: 0, 1, 5, 7...
# reset_index() renumbers from 0 to n in order
# drop=True means don't add old index as a column
```

```python
# ── FINAL SHAPE ──────────────────────────────────────────────
print("\n--- FINAL CLEANED TABLE ---")
print(f"Rows:    {master_clean.shape[0]:,}")
print(f"Columns: {master_clean.shape[1]}")
print("\n--- Remaining Nulls ---")
print(master_clean.isnull().sum()[master_clean.isnull().sum() > 0])
```
