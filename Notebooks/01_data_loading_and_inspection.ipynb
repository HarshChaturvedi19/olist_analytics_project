```python
## code to inspect the csv files 

# a python function that will give all th information of the data frame in one go, this will be very useful for us to quickly understand the data frames we are working with and to identify any issues such as missing values or incorrect data types.
def inspect_df(df,name):
    """ 
    A reusable function to quickly inspect any DataFrame
    In this function, we will print the shape, data types, missing values, and a sample of the data
    """

    print(f"\n{'='*50}")
    print(f" DATASET: {name}")
    print(f"{'='*50}")
    print(f"Shape  : {df.shape[0]} rows * {df.shape[1]} columns")
    print(f"\nColumn Names :\n{list(df.columns)}")
    print(f"\nData Types :\n{df.dtypes}")
    print(f"\nMissing Values :\n{df.isnull().sum()}")
    print(f"\nFirst 5 Rows :\n{df.head()}")


# Now we can use this function to inspect each of our data frames
```

# Table Wise Data Analysis 

## Table 1 - Orders

```python
# Table 1 : Orders
# this is the most important table in our analysis as it contains the main information about the orders, such as the order id, customer id, order status, order purchase timestamp, etc. we will use this table to analyze the order patterns, customer behavior, and to link it with other tables such as order items and payments.
# every other table connects back to this table through the order_id column, so it is crucial to understand this table well.

print("=" * 50)
print("TABLE 1: Orders")
print("=" * 50)

# step 1 : First look 
print("\n --- First 5 Rows ---")
print(orders.head())
# .head() => show first 5 rows of the data frame
```

```python
# step 2 : shape of the data frame
print("\n --- Shape (Rows, Columns) ---")
print(orders.shape)
```

```python
# step 3 : data types of each column
print("\n --- Data Types ---")
print(orders.dtypes)
```

```python
# here we can see that the order_id and customer_id are of type object, which means they are strings. the order_purchase_timestamp is of type object as well, which means it is also a string. we will need to convert this column to a datetime type later on for our analysis. the other columns are of type object as well, which means they are also strings. we will need to convert the order_status column to a categorical type later on for our analysis.
# we need to conver the 
"""
order_purchase_timestamp
order_approved_at  
order_delivered_carrier_date 
order_delivered_customer_date 
order_estimated_delivery_date 
"""
# in to date time format 
date_columns = [
    'order_purchase_timestamp',      # When customer placed the order
    'order_approved_at',             # When payment was approved
    'order_delivered_carrier_date',  # When handed to shipping carrier
    'order_delivered_customer_date', # When customer received it
    'order_estimated_delivery_date'  # Expected delivery date
]

for col in date_columns:
    orders[col] = pd.to_datetime(orders[col])

print("Date columns converted successfully : \n")
print(orders.dtypes)
```

```python
# finding the null values 
print("\n -- NUll Values Per Column ---")
null_counts = orders.isnull().sum()
null_percent = (orders.isnull().sum() / len(orders)) * 100

# create a data frame to access 

null_df = pd.DataFrame({
    'Null Count ': null_counts,
    'Null Percent (%)' : null_percent
})

print(null_df)
```

```python
# See the duplicate Rows
print("\n --- Duplicate Rows ---")
duplicates = orders.duplicated().sum()

print(f"Total Duplicate rows : {duplicates}")
```

```python
# now see the statistical summary 
print("\n --- Statistical Summary ---")
print(orders.describe())
```

```python
# find the unique values in key column ( order_id)
print("\n --- Unique Orders IDs ---")
print(orders['order_id'].nunique())
```

```python
# Order Status Values 
print("\n --- Order Status Values --- ")
print(orders['order_status'].value_counts())
```

```python
# make one column -> Deliery TIme 
print("\n --- Adding Delivery Time Column ---")
orders['delivery_days'] = (orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']).dt.days


print(orders[['order_id', 'order_purchase_timestamp', 
              'order_delivered_customer_date', 'delivery_days']].head(10))
```

```python
# now can see the delivery days statistics
print("\n --- Delivery Days Statistics --- ")
print(orders['delivery_days'].describe())
```

```python
# order volume by year and month
print("\n --- Orders by Year ---")
orders['order_year'] = orders['order_purchase_timestamp'].dt.year

print(orders['order_year'].value_counts().sort_index())
```

```python
# ── PRICE ANALYSIS ──────────────────────────────────────────
print("\n--- Price Statistics ---")
print(items['price'].describe())
# Shows min, max, average item price
# Watch for: max price being extremely high (outliers)
# Watch for: min price being 0 or negative (data error)
```

```python
print("\n--- Freight (Shipping Cost) Statistics ---")
print(items['freight_value'].describe())
# Shipping cost stats
# Compare this with price to understand shipping-to-price ratio
```

```python
# ── OUTLIER CHECK ────────────────────────────────────────────
print("\n--- Items priced above R$2000 (possible outliers) ---")
expensive = items[items['price'] > 2000]
# Filter: keep only rows where price > 2000
# Square brackets with a condition = filtering rows
print(f"Count: {len(expensive)}")
print(expensive['price'].describe())
```

```python
# ── ITEMS PER ORDER ──────────────────────────────────────────
print("\n--- How many items per order? ---")
items_per_order = items.groupby('order_id')['order_item_id'].max()
# groupby('order_id') = group all rows that belong to the same order
# ['order_item_id'].max() = the maximum item number in that order
# If max item_id = 3, that order had 3 items
print(items_per_order.value_counts())
# Shows: how many orders had 1 item, 2 items, 3 items, etc.
```
